```python
import numpy as np
import pandas as pd
```

```python
ntrials = 0
N = [4, 5] # quantifier
target_colors = ["red", "blue", "green"]
filler_colors = ["red", "blue", "green", "purple", "yellow", "black", "gray"]
```

```python
trials = []

for n in N:
    #n_target = [target for target in range(n-3, n+4)]
    print('------- n = ' + str(n))
    for target in range(n-3, n+4):
        
        if (target < n): ntrials = 2
        else: ntrials = 4
        print('ntrials = ' + str(ntrials))
        
        for i_trial in range(ntrials):
            print('* i_trial = ' + str(i_trial))
            
            i_target_color = np.random.randint(len(target_colors))
            target_color = target_colors[i_target_color]
            print('target_color = ' + target_color)
            
            local_filler_colors = filler_colors[0:i_target_color] + filler_colors[i_target_color+1:]
            filler_color = local_filler_colors[np.random.randint(len(local_filler_colors))]            
            print('filler_color = ' + filler_color)
            
            if (target == 1):
                t1 = 1
                t2 = np.random.randint(1, 7)
                f1 = np.random.randint(1, 7)
                f2 = np.random.randint(1, 7)
                while(f1 + f2 + t2 < 12 - target or f1 + f2 + t2 > 16 - target):
                    t2 = np.random.randint(1, 7)
                    f1 = np.random.randint(1, 7)
                    f2 = np.random.randint(1, 7)
                
            else:
                
                t1 = np.random.randint(1, target)
                t2 = target - t1
                while(t1 == t2 and target != 2):
                    t1 = np.random.randint(1, target)
                    t2 = target - t1

                f1 = np.random.randint(1, 7)
                f2 = np.random.randint(1, 7)
                while(f1 + f2 < 12 - target or f1 + f2 > 16 - target):
                    f1 = np.random.randint(1, 7)
                    f2 = np.random.randint(1, 7)
                    
            trials.append([t1, t2, f1, f2, target_color, filler_color])
            
```
